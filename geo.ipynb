```python
import folium
from geopy.geocoders import Nominatim

data = {'country': ['United States', 'Canada', 'France'], 'traffic': [500000, 300000, 200000]}
geolocator = Nominatim(user_agent="geoapi")

map = folium.Map(location=[20, 0], zoom_start=2)

for i, row in enumerate(data['country']):
    location = geolocator.geocode(row)
    folium.CircleMarker(
        location=[location.latitude, location.longitude],
        radius=10,
        popup=f"{row}: {data['traffic'][i]}",
        color="blue",
        fill=True,
    ).add_to(map)

map.save("map.html")
map
```

```python
import pandas as pd

def visualize_similarity_pandas(similarity_data, countries, clusters):

    for region, country_list in countries.items():
        print(f"\n--- {region.upper()} ---")
        for category in clusters:
            print(f"\n**{category}**")
            if region not in similarity_data or category not in similarity_data[region]:
                print("Data not available for this region and category.")
                continue

            table_data = similarity_data[region][category]

            # Create a DataFrame
            df_data = {}
            for country1 in country_list:
                df_data[country1] = {}
                for country2 in country_list:
                    if country1 == country2:
                        df_data[country1][country2] = "-"
                    else:
                        pair = tuple(sorted([country1, country2]))
                        pair_str = "-".join(pair)
                        df_data[country1][country2] = table_data.get(pair_str, "N/A")

            df = pd.DataFrame(df_data)

            # Display the DataFrame (nicely formatted output)
            print(df.to_string()) # to_string() prevents truncation

visualize_similarity_pandas(similarity_data, countries, clusters)
```

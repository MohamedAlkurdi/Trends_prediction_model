```python
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.metrics import mean_absolute_error
import numpy as np
import matplotlib.pyplot as plt
```

```python

cointry_data1 = pd.read_csv("../Classification/output/regions/north_america_australia/genral_labeled_data_with_relative_traffic_rates/Entertainment/USA_with_relative_traffic_rates.csv")
cointry_data2 = pd.read_csv("../Classification/output/regions/north_america_australia/genral_labeled_data_with_relative_traffic_rates/Entertainment/Canada_with_relative_traffic_rates.csv")
cointry_data3 = pd.read_csv("../Classification/output/regions/north_america_australia/genral_labeled_data_with_relative_traffic_rates/Entertainment/Australia_with_relative_traffic_rates.csv")

cointry_data4 = pd.read_csv("../Classification/output/regions/africa/genral_labeled_data_with_relative_traffic_rates/Entertainment/Kenya_with_relative_traffic_rates.csv")
cointry_data5 = pd.read_csv("../Classification/output/regions/africa/genral_labeled_data_with_relative_traffic_rates/Entertainment/Nigeria_with_relative_traffic_rates.csv")
cointry_data6 = pd.read_csv("../Classification/output/regions/africa/genral_labeled_data_with_relative_traffic_rates/Entertainment/SouthAfrica_with_relative_traffic_rates.csv")

cointry_data7 = pd.read_csv("../Classification/output/regions/europe/genral_labeled_data_with_relative_traffic_rates/Entertainment/Denmark_with_relative_traffic_rates.csv")
cointry_data8 = pd.read_csv("../Classification/output/regions/europe/genral_labeled_data_with_relative_traffic_rates/Entertainment/Finland_with_relative_traffic_rates.csv")
cointry_data9 = pd.read_csv("../Classification/output/regions/europe/genral_labeled_data_with_relative_traffic_rates/Entertainment/UK_with_relative_traffic_rates.csv")

cointry_data10 = pd.read_csv("../Classification/output/regions/west_asia/genral_labeled_data_with_relative_traffic_rates/Entertainment/Malaysia_with_relative_traffic_rates.csv")
cointry_data11 = pd.read_csv("../Classification/output/regions/west_asia/genral_labeled_data_with_relative_traffic_rates/Entertainment/Philippines_with_relative_traffic_rates.csv")
cointry_data12 = pd.read_csv("../Classification/output/regions/west_asia/genral_labeled_data_with_relative_traffic_rates/Entertainment/Singapore_with_relative_traffic_rates.csv")
```

```python

def preprocess_data(data):
    data = data.set_index('date')
    data = data['traffic_rate']
    data.index = pd.to_datetime(data.index) 
    X_train = data.loc[data.index < '2017-02-15']
    y_train = data.loc[data.index >= '2017-02-15']
    return X_train, y_train
```

```python
def create_model():
    model = Sequential()
    model.add(LSTM(50, activation='relu', input_shape=(30, 1)))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')
    return model
```

```python

X_train_country1, y_train_country1 = preprocess_data(cointry_data1)
print("X_train_country1")
print(X_train_country1)
print("--------------------------")
print("y_train_country1")
print(y_train_country1)
model = create_model()
model.fit(X_train_country1, y_train_country1, epochs=50, verbose=1)

X_train_country2, y_train_country2 = preprocess_data(cointry_data2)
model.fit(X_train_country2, y_train_country2, epochs=30, verbose=1)

X_train_country3, y_train_country3 = preprocess_data(cointry_data3)
model.fit(X_train_country3, y_train_country3, epochs=30, verbose=1)

X_train_country4, y_train_country4 = preprocess_data(cointry_data4)
model.fit(X_train_country4, y_train_country4, epochs=30, verbose=1)

X_train_country5, y_train_country5 = preprocess_data(cointry_data5)
model.fit(X_train_country5, y_train_country5, epochs=30, verbose=1)

X_train_country6, y_train_country6 = preprocess_data(cointry_data6)
model.fit(X_train_country6, y_train_country6, epochs=30, verbose=1)

X_train_country7, y_train_country7 = preprocess_data(cointry_data7)
model.fit(X_train_country7, y_train_country7, epochs=30, verbose=1)

X_train_country8, y_train_country8 = preprocess_data(cointry_data8)
model.fit(X_train_country8, y_train_country8, epochs=30, verbose=1)

X_train_country9, y_train_country9 = preprocess_data(cointry_data9)
model.fit(X_train_country9, y_train_country9, epochs=30, verbose=1)

X_train_country10, y_train_country10 = preprocess_data(cointry_data10)
model.fit(X_train_country10, y_train_country10, epochs=30, verbose=1)

X_train_country11, y_train_country11 = preprocess_data(cointry_data11)
model.fit(X_train_country11, y_train_country11, epochs=30, verbose=1)

X_train_country12, y_train_country12 = preprocess_data(cointry_data12)
model.fit(X_train_country12, y_train_country12, epochs=30, verbose=1)

```

```python

X_test, y_test = preprocess_data(cointry_data4)
predictions = model.predict(X_test)
mae = mean_absolute_error(y_test, predictions)
print(f"MAE: {mae}")
```

```python
def test_on_time_range(model, data, start_date, end_date, window_size=30):
    filtered_data = data.loc[(data.index >= start_date) & (data.index <= end_date)]
    
    if len(filtered_data) < window_size:
        raise ValueError("Not enough data in the selected time range for testing.")
    
    sequences = []
    for i in range(len(filtered_data) - window_size):
        sequences.append(filtered_data.iloc[i:i+window_size].values)
    X_test = np.array(sequences)
    
    predictions = model.predict(X_test)
    return predictions, filtered_data[window_size:]
```

```python
predictions, actual_data = test_on_time_range(
    model, 
    cointry_data4.set_index('date')['traffic_rate'], 
    start_date='2016-11-28', 
    end_date='2017-05-04'
)

print("Predictions:", predictions)
print("Actual Data:", actual_data)
```

```python

def plot_predictions(predictions, actual_data, title="Model Predictions vs Actual Data"):
    plt.figure(figsize=(25, 5))
    plt.plot(actual_data.index, actual_data.values, label="Actual Data", color='blue')
    prediction_dates = actual_data.index[:len(predictions)]
    plt.plot(prediction_dates, predictions, label="Predictions", color='red', linestyle='--')
    plt.title(title)
    plt.xlabel("Date")
    plt.ylabel("Traffic Rate")
    plt.xticks( rotation=90)
    plt.legend()
    plt.grid()
    plt.show()


plot_predictions(predictions, actual_data)
```
